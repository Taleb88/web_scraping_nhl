```python
import pandas as pd

# Webpage url                                                                                                               
url = 'https://www.hockey-reference.com/teams/BOS/2023.html'

# Extract tables
dfs = pd.read_html(url)

# Get fourth table                                                                                                           
df = dfs[3]

print(df)
```

```python
# creating excel workbook and saving data to it 
df.to_excel('2023_bruins_scoring_regular_season.xlsx')
```

```python
# save data into csv file
csv_df = pd.read_excel('2023_bruins_scoring_regular_season.xlsx')

csv_df.to_csv('2023_bruins_scoring_regular_season.csv', header=True) # using second row for column names

#new_df = pd.DataFrame(pd.read_csv('2023_bruins_scoring_regular_season.csv'))
```

```python
# sorted values via blocks
blocks_df = csv_df.sort_values(by='BLK', ascending=False)
```

```python
# import numpy in order to shift total row to last row position
import numpy as np

bruins_blocked_shots_2023_regular_season_df = blocks_df.iloc[np.arange(-37, len(blocks_df)-37)]
bruins_blocked_shots_2023_regular_season_df
```

```python
# save sorted blocks data to excel workbook
bruins_blocked_shots_2023_regular_season_df.to_excel('2023_bruins_blocks_regular_season.xlsx')
```

```python
# save sorted blocks data to csv
bruins_blocked_shots_2023_regular_season_df.to_csv('2023_bruins_blocks_regular_season.csv', index=False)
```
