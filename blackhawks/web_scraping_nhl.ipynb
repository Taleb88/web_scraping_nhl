```python
import pandas as pd

# Webpage url                                                                                                               
url = 'https://www.hockey-reference.com/teams/CHI/2023.html'

# Extract tables
dfs = pd.read_html(url)

# Get the "Scoring Regular Season" table                                                                                                        
df = dfs[3]

print(df)
```

```python
# creating excel workbook and saving data to it 
df.to_excel('2023_blackhawks_scoring_regular_season.xlsx')
```

```python
# save data into csv file
csv_df = pd.read_excel('2023_blackhawks_scoring_regular_season.xlsx')
```

```python
csv_df
```

```python
csv_df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
csv_df
```

```python
csv_df.to_csv('2023_blackhawks_scoring_regular_season.csv', index=False)
```

```python
# sorted values via blocks
blocks_df = csv_df.sort_values(by='BLK', ascending=False)
```

```python
blocks_df
```

```python
# import numpy in order to shift total row to last row position
import numpy as np

blackhawks_blocked_shots_2023_regular_season_df = blocks_df.iloc[np.arange(-44, len(blocks_df)-44)]
blackhawks_blocked_shots_2023_regular_season_df
```

```python
# save sorted blocks data to excel workbook
blackhawks_blocked_shots_2023_regular_season_df.to_excel('2023_blackhawks_blocks_regular_season.xlsx')
```

```python
# save sorted blocks data to csv
blackhawks_blocked_shots_2023_regular_season_df.to_csv('2023_blackhawks_blocks_regular_season.csv', index=False)
```
